## Divergence of the harmonic series

```python
# Compute several values of the partial sum
n = symbols('n')
for N in range(6):
    Sn = Sum(1/n, (n,1,10**N))
    print('S_', 10**N,' = ', Sn.doit().evalf())
```

```python
Sum(1/n, (n,1,oo)).is_convergent()
```

```python
Sum(1/n, (n,1,oo)).doit()
```

## Plotting a function in SymPy

```python
from sympy import *
x = symbols('x')
f = 1/3*x**3 - 1/2*x**2-x+2
plot(f)
```

```python
plot(f,(x,-2,1))
```

```python
f = 2/sqrt(x-8)
plot(f,(x,8.01,25))
```
