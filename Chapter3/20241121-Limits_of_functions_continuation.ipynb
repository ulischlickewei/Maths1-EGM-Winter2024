## SymPy can compute limits

**Example**: $$\lim_{x\to 0} \frac{\sin(x)}{x}.$$

```python
f = sin(x)/x
limit(f,x,0)
```

**Example**: See that $$\lim_{x\to 0}\sin\left(\frac{\pi}{x}\right)$$
does not exist.

```python
limit(sin(pi/x), x, 0)
```

**Example**: Determine
$$\lim_{x \to \infty} \mathrm{e}^{-x^2}.$$

```python
limit(exp(-x**2), x, oo)
```

**Example**: Determine the limit of the sequence $a_n = \frac{2n+3}{4n+4}$.

```python
n = symbols('n')
an = (2*n+3)/(4*n+4)
limit(an,n,oo)
```

**Example**: Let $b \in \mathbb{R}$. Determine
$$\lim_{n\to \infty} \frac{2b n^2 - 2}{3n^2+2n}.$$

```python
b = symbols('b')
limit((2*b*n**2)/(3*n**2 + 2), n, oo)
```
