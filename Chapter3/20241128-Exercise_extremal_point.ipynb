```python
from sympy import *
```

```python
x = symbols('x')
f = 3*exp(2*x+1) + 6*x*exp(2*x+1)
f
```

```python
fp = diff(f,x)
fp
```

```python
# ATTENTION: solve can miss zeros!!!
solve(fp,x)
```

```python
# To check if critical point is extremal point
plot(f,(x,-1.1,-0.9))
```

```python
diff(f,x,2).subs(x,-1).evalf()
```
