```python
# Compute Taylor polynomial of deg 2
from sympy import *
x = symbols('x')
f = 1/x
series(f,x, 1, 3)
```

```python
f3p = diff(f,x,3)
f3p
```

```python
plot(f3p,(x,0.7,1.3))
```

```python
f3p.subs(x,0.7)
```

```python
f3p.subs(x,0.7)*(0.3)**3/factorial(3)
```
